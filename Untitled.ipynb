```python
import os
import logging
import pandas as pd
import numpy as np

# ========== 1. Configuration and Logging ==========
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s %(levelname)s: %(message)s'
)
logger = logging.getLogger(__name__)

DATA_PATH = os.getenv('HDDM_DATA_PATH', 'trials.csv')
OUTPUT_PATH = 'jin_hddm_ready.csv'
SUMMARY_PATH = 'trial_count_by_condition.csv'
DROPPED_PATH = 'dropped_trials.csv'  # Export dropped trials for traceability

# ========== 2. Load Data ==========
logger.info(f"Loading data: {DATA_PATH}")
df = pd.read_csv(DATA_PATH)

# ========== 3. Extract subj_idx ==========
id_candidates = ['participant_id', 'subject', 'subj_id']
subj_col = next((col for col in id_candidates if col in df.columns), None)
if subj_col:
    df['subj_idx'] = (
        df[subj_col]
          .astype(str)
          .str.extract(r'(\d+)$', expand=False)
          .astype(int)
    )
else:
    raise ValueError("No participant_id/subject/subj_id column found in the input data!")

logger.info(f"Found {df['subj_idx'].nunique()} subjects: {sorted(df['subj_idx'].unique())}")

# ========== 4. Check Missing Values ==========
na_cols = df.columns[df.isna().any()]
if len(na_cols) > 0:
    logger.warning(f"Columns with missing values: {list(na_cols)}. Please check if this is expected.")

# ========== 5. RT and Response Preprocessing ==========
if df['RT'].max() > 20:
    df['rt'] = df['RT'] / 1000.0
else:
    df['rt'] = df['RT']

# Binary response recoding (robust to different encodings)
if df['reaction'].dtype == 'O':
    df['reaction'] = df['reaction'].str.lower().str.strip()
response_map = {1: 1, 2: 0, 'accept': 1, 'reject': 0}
df['response'] = df['reaction'].replace(response_map)
if not set(df['response'].dropna().unique()).issubset({0, 1}):
    logger.warning(f"Unknown values in 'reaction': {df['reaction'].unique()}. Please check!")
    raise ValueError("Response variable must be 0/1!")

# ========== 6. Offer Type Detection ==========
def calc_offer_type(other, you):
    if pd.isna(other) or pd.isna(you):
        return np.nan
    pair = f"{int(other)}:{int(you)}"
    if pair in ["5:5", "6:4"]:
        return "fair"
    if pair in ["8:2", "9:1"]:
        return "unfair"
    return "other"

df['offer_type'] = df.apply(lambda row: calc_offer_type(row['Offers_Other'], row['Offers_You']), axis=1)
main_offer_types = ["fair", "unfair"]
df = df[df['offer_type'].isin(main_offer_types)].copy()
df['offer_type'] = pd.Categorical(df['offer_type'], categories=main_offer_types, ordered=True)

# ========== 7. Categorical Variable Handling ==========
emotion_categories = ["neu", "aff", "dis", "dom", "enj"]
df['emotion'] = pd.Categorical(df['emotion'].str.lower().str.strip(), categories=emotion_categories, ordered=True)
df['offer_type'] = pd.Categorical(df['offer_type'].str.lower().str.strip(), categories=main_offer_types, ordered=True)

# Check for unknown categories
for col, cats in [('emotion', emotion_categories), ('offer_type', main_offer_types)]:
    nan_rate = df[col].isna().mean()
    if nan_rate > 0:
        logger.warning(f"Column {col} has {nan_rate:.1%} unknown or missing categories. Please check the raw data.")

# ========== 8. RT Outlier Removal ==========
df = df.query('0.15 <= rt <= 3.0').copy()
df['rt_z'] = df.groupby('subj_idx')['rt'].transform(lambda x: (x - x.mean()) / x.std(ddof=0))
dropped = df[df['rt_z'].abs() > 3].copy()
df = df.query('rt_z.abs() <= 3').drop(columns='rt_z')
if not dropped.empty:
    dropped.to_csv(DROPPED_PATH, index=False)
    logger.info(f"Dropped trials (by z-score) have been saved to: {DROPPED_PATH}")

# ========== 9. Trial Count and Behavior Overview ==========
summary = (
    df
    .groupby(['subj_idx', 'emotion', 'offer_type', 'response'])
    .size()
    .reset_index(name='trial_n')
)
pivot_summary = summary.pivot_table(
    index=['subj_idx', 'emotion', 'offer_type'],
    columns='response',
    values='trial_n',
    fill_value=0
).reset_index()

pivot_summary = pivot_summary.rename(columns={0: 'reject_n', 1: 'accept_n'})
pivot_summary['total_n'] = pivot_summary['accept_n'] + pivot_summary['reject_n']
pivot_summary['accept_rate'] = np.where(
    pivot_summary['total_n'] > 0,
    pivot_summary['accept_n'] / pivot_summary['total_n'],
    np.nan
)
pivot_summary.to_csv(SUMMARY_PATH, index=False)
logger.info(f"Saved per-subject/condition accept/reject counts to: {SUMMARY_PATH}")

# Optional: Plot cell trial number distribution for quality check
try:
    import matplotlib.pyplot as plt
    pivot_summary['total_n'].hist(bins=20)
    plt.title('Trial count distribution per cell')
    plt.xlabel('Number of trials')
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
except Exception as e:
    logger.warning(f"Plot error (non-fatal): {e}")

print(pivot_summary.head(10))

# ========== 10. Minimum Trial Check ==========
MIN_TRIALS_PER_CELL = 8
if (pivot_summary['total_n'] < MIN_TRIALS_PER_CELL).any():
    low_cell = pivot_summary[pivot_summary['total_n'] < MIN_TRIALS_PER_CELL]
    logger.warning(
        f"Warning: {len(low_cell)} subject-condition cells have fewer than {MIN_TRIALS_PER_CELL} trials! Modeling may be unstable."
    )
    print(low_cell)

# ========== 11. Export Cleaned Data ==========
n_subj, n_trials = df['subj_idx'].nunique(), len(df)
logger.info(f"After cleaning: {n_subj} subjects, {n_trials} trials")
df.to_csv(OUTPUT_PATH, index=False)
logger.info(f"Saved analysis-ready data to: {OUTPUT_PATH}")

# ========== 12. Export README ==========
with open('data_preparation_README.txt', 'w', encoding='utf-8') as f:
    f.write(
        f"""# Data Preparation Log
Raw file: {DATA_PATH}
Subjects: {n_subj}
Total trials: {n_trials}
Emotion categories: {emotion_categories}
Offer types: {main_offer_types}
RT filter: 0.15‚Äì3.0s + within-subject z-score ¬±3
Dropped trials (z): {DROPPED_PATH if not dropped.empty else 'None'}
Output files: {OUTPUT_PATH}, {SUMMARY_PATH}
Last saved: {pd.Timestamp.now()}
"""
    )
```

```python
import hddm
import pandas as pd
import arviz as az
import matplotlib.pyplot as plt
import gc, os, sys
import logging

# ========== Êó•ÂøóÁ≥ªÁªü ==========
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s %(levelname)s: %(message)s'
)
logger = logging.getLogger(__name__)

# ========== Êï∞ÊçÆË∑ØÂæÑËÆæÁΩÆ ==========
DATA_PATH = 'jin_hddm_ready.csv'
if not os.path.exists(DATA_PATH):
    logger.error(f"Êï∞ÊçÆÊñá‰ª∂ {DATA_PATH} ‰∏çÂ≠òÂú®ÔºÅËØ∑ÂÖàËøêË°åÊï∞ÊçÆÂáÜÂ§áËÑöÊú¨„ÄÇ")
    raise SystemExit(1)
else:
    logger.info(f"ÂáÜÂ§áÂä†ËΩΩÊï∞ÊçÆÊñá‰ª∂Ôºö{DATA_PATH}")

# ========== ËØªÂèñÊï∞ÊçÆ ==========
df = pd.read_csv(DATA_PATH)
logger.info(f"Êï∞ÊçÆËØªÂèñÊàêÂäüÔºåtrialÊÄªÊï∞Ôºö{len(df)}ÔºåË¢´ËØïÊï∞Ôºö{df['subj_idx'].nunique()}")

print("Êï∞ÊçÆÂ≠óÊÆµ‰ø°ÊÅØÔºö")
print(df.info())
print("Êï∞ÊçÆÈ¢ÑËßàÔºö\n", df.head())

# ========== ÂÖ≥ÈîÆÂèòÈáèÊ†°È™å ==========
essential_cols = ['subj_idx', 'rt', 'response', 'emotion', 'offer_type']
for col in essential_cols:
    if col not in df.columns:
        logger.error(f"Êï∞ÊçÆÁº∫Â∞ëÂÖ≥ÈîÆÂàó: {col}ÔºåËØ∑Ê£ÄÊü•Êï∞ÊçÆÂáÜÂ§áËøáÁ®ã„ÄÇ")
        raise SystemExit(1)
logger.info("Êï∞ÊçÆÂ≠óÊÆµÊ†°È™åÈÄöËøá„ÄÇ")

# ========== Ê£ÄÊü•ÊûÅÁ´ØÂèçÂ∫îÊó∂ ==========
if df['rt'].min() < 0.15 or df['rt'].max() > 3.0:
    logger.warning(f"ÂèëÁé∞ÊûÅÁ´ØÂèçÂ∫îÊó∂Âå∫Èó¥Ôºö{df['rt'].min()} ~ {df['rt'].max()}sÔºåÂèØËÉΩÊúâÂºÇÂ∏∏trialÊú™ÂâîÈô§„ÄÇ")
else:
    logger.info("ÂèçÂ∫îÊó∂Âå∫Èó¥Ê≠£Â∏∏„ÄÇ")

# ========== ÂàÜÁ±ªÂèòÈáèÂîØ‰∏ÄÂÄºÊ£ÄÊü• ==========
for col in ['emotion', 'offer_type']:
    uniq = set(df[col].unique())
    # Â¶ÇÊûú‰∏∫Categorical, Áõ¥Êé•ÊØîÂØπcategories
    if hasattr(df[col], 'cat'):
        cat_set = set(df[col].cat.categories)
        extra = uniq - cat_set
        if extra:
            logger.warning(f"{col}ÂàóÂ≠òÂú®Êú™Ê≥®ÂÜåÁ±ªÂà´: {extra}")

# ========== ËæìÂá∫Êï∞ÊçÆÁªìÊûÑÈÄüËßà ==========
logger.info(f"\nÂêÑÊù°‰ª∂trialÂàÜÂ∏ÉÁªüËÆ°ÔºàË°®Ê†ºÔºâÔºö\n{df.groupby(['emotion','offer_type']).size().unstack(fill_value=0)}")

# ========== ÂçèÂèòÈáèÈ¢ÑËßàÔºàÂ¶ÇÂ≠òÂú®Ôºâ ==========
for cov in ['FRN_z', 'LPP_offer_z']:
    if cov in df.columns:
        logger.info(f"{cov} ÂçèÂèòÈáèÈÄüËßàÔºöÂùáÂÄº={df[cov].mean():.3f}, ÊñπÂ∑Æ={df[cov].std():.3f}, Áº∫Â§±={df[cov].isna().sum()}")

# ========== ÈáäÊîæÂÜÖÂ≠ò ==========
gc.collect()
```

üö©Chunk 2ÔºöÊ®°ÂûãÊâπÈáèÈÖçÁΩÆÔºàÂèÇÊï∞„ÄÅÂÖ¨Âºè„ÄÅ‰øùÂ≠òË∑ØÂæÑÔºâ

```python
import os
import gc
import logging
import pandas as pd
import hddm

# ===== 1. Config =====
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s %(levelname)s: %(message)s'
)
logger = logging.getLogger(__name__)

DATA_PATH = 'jin_hddm_ready.csv'
BASE_OUTDIR = 'model_results'
os.makedirs(BASE_OUTDIR, exist_ok=True)

MODEL_CONFIGS = [
#     # 1. Âü∫Á∫øÊ®°ÂûãÔºàÂÖ®‰∏çÂèò/null modelÔºâ‚Äî‚ÄîÊâÄÊúâÂèÇÊï∞Âùá‰∏∫Êà™Ë∑ùÔºå‰∏çÈöèÊù°‰ª∂ÂèòÂåñ
#     {
#         'name': 'null_model',
#         'formula': [
#             "v ~ 1",
#             "a ~ 1"
#             # Ëã•ÈúÄË¶Åt, zÂêåÁêÜÊ∑ªÂä†Ôºö"t ~ 1", "z ~ 1"
#         ],
#         'desc': 'Null model: all parameters are intercept-only (no condition effects)'
#     },
    # 2. ÂÖ®ÂèòÊ®°ÂûãÔºàvÂíåaÈÉΩÂèò/fully flexible modelÔºâ‚Äî‚ÄîvÂíåaÈÉΩÈöèÊù°‰ª∂ÂèòÂåñ‰∏îÊúâ‰∫§‰∫í
    {
        'name': 'full_model',
        'formula': [
            "v ~ 1 + C(emotion, Sum) * C(offer_type, Sum)",
            "a ~ 1 + C(emotion, Sum) * C(offer_type, Sum)"
            # t, zÂèÇÊï∞‰∏ÄËà¨‰∏çÂª∫ËÆÆÂä†‰∫§‰∫íÔºåÈÄöÂ∏∏‰øùÂÆàÂä†Ôºö"t ~ 1", "z ~ 1"
        ],
        'desc': 'Full model: v and a both vary with emotion, offer_type, and their interaction'
    }
#     # 3. ‰ªÖvÂèò‰∏ªÊïàÂ∫îÊ®°Âûã
#     {
#         'name': 'v_main',
#         'formula': [
#             "v ~ 1 + C(emotion, Sum) + C(offer_type, Sum)",
#             "a ~ 1"
#         ],
#         'desc': 'Drift rate (v) main effects: emotion and offer_type (a only intercept)'
#     },
#     # 4. ‰ªÖvÂèò‰∫§‰∫íÊ®°Âûã
#     {
#         'name': 'v_interaction',
#         'formula': [
#             "v ~ 1 + C(emotion, Sum) * C(offer_type, Sum)",
#             "a ~ 1"
#         ],
#         'desc': 'Drift rate (v) interaction: emotion √ó offer_type (a only intercept)'
#     },
#     # 5. ‰ªÖaÂèò‰∏ªÊïàÂ∫îÊ®°Âûã
#     {
#         'name': 'a_main',
#         'formula': [
#             "v ~ 1",
#             "a ~ 1 + C(emotion, Sum) + C(offer_type, Sum)"
#         ],
#         'desc': 'Boundary separation (a) main effects: emotion and offer_type (v only intercept)'
#     },
#     # 6. ‰ªÖaÂèò‰∫§‰∫íÊ®°Âûã
#     {
#         'name': 'a_interaction',
#         'formula': [
#             "v ~ 1",
#             "a ~ 1 + C(emotion, Sum) * C(offer_type, Sum)"
#         ],
#         'desc': 'Boundary separation (a) interaction: emotion √ó offer_type (v only intercept)'
#     }
]


INCLUDE_PARAMS = ['a', 'v', 't', 'z']
N_SAMPLES = 8000
N_BURN    = 1000
N_CHAINS  = 4

# ===== 2. Data Loading & Checking =====
if not os.path.exists(DATA_PATH):
    logger.error(f"Data file {DATA_PATH} does not exist!")
    raise SystemExit(1)

df = pd.read_csv(DATA_PATH)
logger.info(f"Data loaded: {len(df)} trials, {df['subj_idx'].nunique()} subjects.")

for col in ['subj_idx', 'rt', 'response', 'emotion', 'offer_type']:
    if col not in df.columns:
        logger.error(f"Missing required column: {col}")
        raise SystemExit(1)
assert set(df['response'].unique()).issubset({0, 1}), "Response variable must be 0/1!"

# ===== 3. Model Fitting & .hddm Saving =====
for cfg in MODEL_CONFIGS:
    logger.info(f"======= Start fitting: {cfg['name']} =======")
    logger.info(f"Model: {cfg['desc']}\nFormula: {cfg['formula']}")
    outdir = os.path.join(BASE_OUTDIR, cfg['name'])
    os.makedirs(outdir, exist_ok=True)

    model = hddm.HDDMRegressor(
        df,
        cfg['formula'],
        include=INCLUDE_PARAMS,
        group_only_regressors=False,  # subject-level variance (hierarchical)
        is_group_model=True,
        p_outlier=0.05,
        keep_regressor_trace=False,
        informative=False
    )
    logger.info(f"Sampling with {N_CHAINS} chains, {N_SAMPLES} samples, {N_BURN} burn-in ...")
    model.sample(N_SAMPLES, burn=N_BURN, chains=N_CHAINS)
    modelfile = os.path.join(outdir, f"{cfg['name']}.hddm")
    model.save(modelfile)
    logger.info(f"Model saved: {modelfile}")

    # ==== Âä† priorËá™Âä®ÈááÊ†∑‰∏é‰øùÂ≠ò ====
    n_prior = 3000  # priorÊ†∑Êú¨Êï∞ÔºåÂèØËá™ÂÆö‰πâ
    prior_dict = {}
    for param in INCLUDE_PARAMS:
        prior_dict[param] = np.random.normal(0, 1, n_prior)
    prior_df = pd.DataFrame(prior_dict)
    prior_path = os.path.join(outdir, f"{cfg['name']}_prior_samples.csv")
    prior_df.to_csv(prior_path, index=False)
    print(f"Prior samples for {cfg['name']} saved: {prior_path}")
    # ÂÜÖÂ≠òÊ∏ÖÁêÜ
    del model
    gc.collect()
    logger.info(f"======= {cfg['name']} fitting done, memory released =======")

logger.info("All models and chains finished successfully!")
```

üö©Chunk 3.1ÔºöHDDMÊ®°ÂûãÊâπÈáèÂØºÂá∫InferenceDataÔºàNetCDF, .ncÔºâ

```python
import os
import gc
import hddm
import logging
from tqdm import tqdm

# ========== Âü∫Êú¨ÈÖçÁΩÆ ==========
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s %(levelname)s: %(message)s'
)
logger = logging.getLogger(__name__)

MODEL_CONFIGS = [
    {'name': 'full_model'},
]
BASE_OUTDIR = "model_results"

def gen_and_save_infdata_fast(model_name, model_dir):
    logger.info(f"\n==== [Group-level InferenceData (ppc)] {model_name} ====")
    hddm_path = os.path.join(model_dir, f"{model_name}.hddm")
    if not os.path.exists(hddm_path):
        logger.warning(f"‚ùå {hddm_path} not found, skipping.")
        return
    try:
        model = hddm.load(hddm_path)
        nc_path = os.path.join(model_dir, f"{model_name}_grouponly.nc")
        # ÂØºÂá∫NetCDFÔºåÂåÖÂê´ÂêéÈ™åÈ¢ÑÊµãÔºàppcÔºâ„ÄÅÂÖàÈ™å
        model.to_infdata(
            loglike=False,
            ppc=True,
            sample_prior=True,
            save_name=nc_path
        )
        logger.info(f"InferenceData saved as {nc_path}")
        del model
        gc.collect()
    except Exception as e:
        logger.error(f"‚ùå Failed to create InferenceData: {model_name}: {e}")

# ========== ‰∏ªÂæ™ÁéØ ==========
for cfg in tqdm(MODEL_CONFIGS, desc='Exporting group-level InferenceData'):
    model_dir = os.path.join(BASE_OUTDIR, cfg['name'])
    gen_and_save_infdata_fast(cfg['name'], model_dir)

print("All group-level InferenceData exported.")
```

```python
import arviz as az
import numpy as np
import pandas as pd
import os
import shutil

# 1. ËØªÂèñË°å‰∏∫Êï∞ÊçÆÔºåÁ°Æ‰øùÊ†∏ÂøÉÂàóÊó†NA
df = pd.read_csv("jin_hddm_ready.csv")
behavior_cols = ['rt', 'response', 'emotion', 'offer_type']
df = df.dropna(subset=behavior_cols).reset_index(drop=True)
print(f"È¢ÑÂ§ÑÁêÜÂêéÂâ©‰Ωô trial Êï∞: {len(df)}")

# 2. ËÆæÁΩÆÊ®°Âûã‰∏éÊñá‰ª∂Ë∑ØÂæÑ
MODEL_NAMES = ['full_model']
BASE_OUTDIR = "model_results"

for model_name in MODEL_NAMES:
    model_dir = os.path.join(BASE_OUTDIR, model_name)
    nc_in = os.path.join(model_dir, f"{model_name}_grouponly.nc")
    nc_tmp = os.path.join(model_dir, f"{model_name}_grouponly_obs.nc")
    if not os.path.exists(nc_in):
        print(f"{nc_in} ‰∏çÂ≠òÂú®ÔºåË∑≥Ëøá")
        continue

    inf = az.from_netcdf(nc_in)
    print(f"{model_name} Groups: {inf.groups()}")

    # Ê£ÄÊü•posterior_predictiveÂàÜÁªÑÔºåÂä®ÊÄÅË°•ÈΩê observed_data
    if hasattr(inf, "posterior_predictive"):
        observed_dict = {}
        # Âä®ÊÄÅÊ£ÄÊµãÊâÄÊúâppc‰∏≠ÁöÑkey
        for key in inf.posterior_predictive.keys():
            if key in df.columns:
                observed_dict[key] = df[key].values
            else:
                print(f"Ë≠¶ÂëäÔºö{key} ‰∏çÂú®dfÂàó‰∏≠ÔºåË∑≥Ëøá")
        inf2 = inf.copy()
        inf2.add_groups({"observed_data": observed_dict})

        # 1. ÂÖà‰øùÂ≠ò‰∏∫‰∏¥Êó∂Êñ∞Êñá‰ª∂
        az.to_netcdf(inf2, nc_tmp)
        print(f"{model_name} Â∑≤Ë°•ÂÖ® observed_dataÔºåÂè¶Â≠ò‰∏∫: {nc_tmp}")

        # 2. Ê£ÄÊü•Êó†ËØØÂêéÔºå**ÂÆâÂÖ®Ë¶ÜÁõñÂéüÊñá‰ª∂**
        try:
            az.from_netcdf(nc_tmp)  # Â∞ùËØïËØªÂèñÔºåÁ°ÆËÆ§Êñá‰ª∂Êó†ËØØ
            shutil.move(nc_tmp, nc_in)
            print(f"{model_name} Êñá‰ª∂Â∑≤ÂÆâÂÖ®Ë¶ÜÁõñ‰∏∫Ê†áÂáÜÁªìÊûÑÔºö{nc_in}")
        except Exception as e:
            print(f"Ë¶ÜÁõñÂ§±Ë¥•ÔºåÊú™ÊõøÊç¢ÂéüÊñá‰ª∂: {e}")
    else:
        print(f"{model_name} Ê≤°Êúâ posterior_predictiveÔºåË∑≥Ëøá observed_data Ë°•ÈΩê")

print("ÂÖ®ÈÉ® observed_data Ë°•ÈΩêÂπ∂ÂÆâÂÖ®Ë¶ÜÁõñÂéüNetCDFÊñá‰ª∂Â∑≤ÂÆåÊàê„ÄÇ")
```

```python
import os
import gc
import logging
import pandas as pd
import hddm

logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s %(levelname)s: %(message)s'
)
logger = logging.getLogger(__name__)

BASE_OUTDIR = 'model_results'
MODEL_CONFIGS = [
    {'name': 'full_model'},
]

for cfg in MODEL_CONFIGS:
    model_dir = os.path.join(BASE_OUTDIR, cfg['name'])
    modelfile = os.path.join(model_dir, f"{cfg['name']}.hddm")
    if not os.path.exists(modelfile):
        logger.warning(f"Model file not found: {modelfile}")
        continue

    logger.info(f"Loading model: {modelfile}")
    model = hddm.load(modelfile)

    # Export parameter summary (group + subject)
    try:
        summary_df = model.gen_stats()
        summary_path = os.path.join(model_dir, f"{cfg['name']}_summary.csv")
        summary_df.to_csv(summary_path)
        logger.info(f"Parameter summary exported to: {summary_path}")
    except Exception as e:
        logger.warning(f"Failed to export parameter summary: {e}")

    # Export full MCMC traces (all chains together)
    try:
        trace = model.get_traces()
        trace_path = os.path.join(model_dir, f"{cfg['name']}_trace.csv")
        trace.to_csv(trace_path)
        logger.info(f"Trace exported to: {trace_path}")
    except Exception as e:
        logger.warning(f"Failed to export trace: {e}")

    del model
    gc.collect()
    logger.info(f"Diagnostics done for {cfg['name']}")

print("All models' summaries and traces exported (multi-chain version).")
```

```python
import arviz as az
import matplotlib.pyplot as plt
import os
import pandas as pd

# ========== 1. Âü∫Êú¨ÈÖçÁΩÆ ==========
BASE_OUTDIR = "model_results"
MODEL_CONFIGS = [
    {'name': 'v_main'},
    {'name': 'v_interaction'},
    {'name': 'a_main'},
    {'name': 'a_interaction'},
    {'name': 'full_model'},
]
diagnostic_dir = os.path.join(BASE_OUTDIR, "_diagnostics")
os.makedirs(diagnostic_dir, exist_ok=True)

RHAT_THRESH = 1.01  # Êé®ËçêÊ†áÂáÜÔºå‰∏•Ê†ºÂÆ°Á®øÂèØÁî®1.01

# ‰Ω†ÊúÄÂÖ≥ÂøÉÁöÑ‰∏ªÁªìËÆ∫ÂèÇÊï∞ÔºàÁ§∫‰æãÔºåÂèØÂä®ÊÄÅÊõ¥Êñ∞Ôºâ
MAIN_EFFECTS = [
    "v_C(emotion, Sum)[S.aff]",
    "v_C(emotion, Sum)[S.dis]",
    "v_C(emotion, Sum)[S.dom]",
    "v_C(emotion, Sum)[S.enj]",
    "v_C(offer_type, Sum)[S.unfair]",
    "v_C(emotion, Sum)[S.aff]:C(offer_type, Sum)[S.unfair]",
    "v_C(emotion, Sum)[S.dis]:C(offer_type, Sum)[S.unfair]",
    "v_C(emotion, Sum)[S.dom]:C(offer_type, Sum)[S.unfair]",
    "v_C(emotion, Sum)[S.enj]:C(offer_type, Sum)[S.unfair]",
    "a_C(emotion, Sum)[S.aff]",
    "a_C(emotion, Sum)[S.dis]",
    "a_C(emotion, Sum)[S.dom]",
    "a_C(emotion, Sum)[S.enj]",
    "a_C(offer_type, Sum)[S.unfair]",
    "a_C(emotion, Sum)[S.aff]:C(offer_type, Sum)[S.unfair]",
    "a_C(emotion, Sum)[S.dis]:C(offer_type, Sum)[S.unfair]",
    "a_C(emotion, Sum)[S.dom]:C(offer_type, Sum)[S.unfair]",
    "a_C(emotion, Sum)[S.enj]:C(offer_type, Sum)[S.unfair]",
]

# ========== 2. TraceplotÁªòÂà∂ÂáΩÊï∞ ==========
def plot_trace_with_smart_font(inf, params, save_path):
    """ÊâπÈáètraceplotËá™Âä®ÁæéÂåñÔºàÂª∫ËÆÆÂè™Áîªgroup-levelÂèÇÊï∞Ôºâ"""
    if not params:
        print("No params to plot.")
        return
    n_params = len(params)
    figsize = (12, max(2*n_params, 8))
    axes = az.plot_trace(
        inf, var_names=params, compact=False,
        legend=True, figsize=figsize
    )
    if hasattr(axes, 'flat'):
        axes = axes.flat
    for ax in axes:
        try:
            ax.title.set_fontsize(10)
            ax.xaxis.label.set_fontsize(9)
            ax.yaxis.label.set_fontsize(9)
            for label in (ax.get_xticklabels() + ax.get_yticklabels()):
                label.set_fontsize(8)
        except Exception:
            pass
    plt.tight_layout()
    plt.savefig(save_path)
    plt.close()
    print(f"Traceplot saved: {save_path}")

# ========== 3. Â∑•ÂÖ∑ÂáΩÊï∞ ==========
def is_group(idx):
    """‰ªÖÂåÖÂê´ÁªÑÊ∞¥Âπ≥ÂèÇÊï∞Ôºà‰∏çÂê´subj/subjectÁ≠â‰∏™‰ΩìÂèÇÊï∞Ôºâ"""
    return ('subj' not in idx.lower()) and ('subject' not in idx.lower())

# ========== 4. ‰∏ªËØäÊñ≠Âæ™ÁéØ ==========
summary_all = []
group_level_status = {}
all_bad_rhat_params = []

for cfg in MODEL_CONFIGS:
    name = cfg['name']
    # Êô∫ËÉΩËØÜÂà´Â§öÁßçncÂëΩÂêç
    nc_path_candidates = [
        os.path.join(BASE_OUTDIR, name, f"{name}_allchains.nc"),
        os.path.join(BASE_OUTDIR, name, f"{name}.nc"),
        os.path.join(BASE_OUTDIR, name, f"{name}_grouponly.nc"),
    ]
    for nc_path in nc_path_candidates:
        if os.path.exists(nc_path):
            break
    else:
        print(f"‚ùó Missing model: {name}")
        continue

    inf = az.from_netcdf(nc_path)

    # 1. ËØäÊñ≠ÊåáÊ†áÔºàÊâÄÊúâÂèÇÊï∞Ôºâ
    diag = az.summary(inf, kind='diagnostics', round_to=6)
    diag_path = os.path.join(diagnostic_dir, f'{name}_diagnostics.csv')
    diag.to_csv(diag_path)
    print(f"{name} diagnostics saved: {diag_path}")

    # 2. ‰ªÖÁªÑÊ∞¥Âπ≥ÂèÇÊï∞
    group_params = [idx for idx in diag.index if is_group(idx)]
    # ÁªÑÊ∞¥Âπ≥Êî∂ÊïõÊÄß
    bad_group_rhat = diag.loc[group_params][diag.loc[group_params, 'r_hat'] > RHAT_THRESH]
    if not bad_group_rhat.empty:
        bad_group_path = os.path.join(diagnostic_dir, f'{name}_bad_group_rhat.csv')
        bad_group_rhat.to_csv(bad_group_path)
        print(f"‚ö†Ô∏è {name}: Non-converged group-level params (Rhat > {RHAT_THRESH}): {bad_group_path}")
        group_level_status[name] = "‚ùåNon-converged"
        all_bad_rhat_params.append(bad_group_rhat.assign(model=name))
    else:
        print(f"{name}: All group-level params converged (Rhat ‚â§ {RHAT_THRESH})!")
        group_level_status[name] = "‚úÖAll converged"

    # 4. ‰∏ªÁªìËÆ∫ÂèÇÊï∞Êî∂ÊïõÊÄß
    main_rhat = diag.loc[diag.index.intersection(MAIN_EFFECTS), 'r_hat']
    main_bad = main_rhat[main_rhat > RHAT_THRESH]
    if not main_bad.empty:
        print(f"‚ùå {name}: Not converged MAIN_EFFECTS: {main_bad.to_dict()}")
    else:
        print(f"‚úÖ {name}: All MAIN_EFFECTS parameters converged (r_hat ‚â§ {RHAT_THRESH})")

    # 5. ÂèÇÊï∞ÂêéÈ™åÊëòË¶ÅÔºàÁªÑÊ∞¥Âπ≥+È´òÁ≤æÂ∫¶r_hatÔºâ
    summary = az.summary(inf, round_to=4)
    summary['model'] = name
    # Áî®ËØäÊñ≠Ëé∑ÂæóÁöÑÈ´òÁ≤æÂ∫¶r_hatË¶ÜÁõñ
    summary['r_hat'] = diag.loc[summary.index, 'r_hat']
    summary_path = os.path.join(diagnostic_dir, f'{name}_posterior_summary.csv')
    summary.to_csv(summary_path, encoding='utf-8-sig')
    print(f"{name} posterior summary saved: {summary_path}")

    summary_all.append(summary)

    # 6. ÁªòÂà∂ÊâÄÊúâgroup-levelÂèÇÊï∞ÁöÑtraceplot
    group_params_plot = [v for v in inf.posterior.data_vars if is_group(v)]
    trace_save_path = os.path.join(diagnostic_dir, f"{name}_traceplot.pdf")
    plot_trace_with_smart_font(inf, group_params_plot, trace_save_path)

# 7. Êî∂ÊïõÊÄß‰∏ÄËßàË°®
status_df = pd.DataFrame({
    'model': list(group_level_status.keys()),
    'group_level_rhat_status': list(group_level_status.values())
})
status_path = os.path.join(diagnostic_dir, "group_level_rhat_status.csv")
status_df.to_csv(status_path, index=False, encoding='utf-8-sig')
print(f"\nAll model group-level convergence status saved: {status_path}")

# 8. ÂêàÂπ∂ÊâÄÊúâÊ®°ÂûãÂêéÈ™åÊëòË¶Å
if summary_all:
    summary_all_df = pd.concat(summary_all, keys=[s['model'][0] for s in summary_all])
    all_summary_path = os.path.join(diagnostic_dir, "all_models_posterior_summary.csv")
    summary_all_df.to_csv(all_summary_path, encoding='utf-8-sig')
    print(f"All models' posterior summaries saved: {all_summary_path}")

# 9. Ê±áÊÄªÊâÄÊúâ‰∏çÊî∂ÊïõÂèÇÊï∞
if all_bad_rhat_params:
    all_bad_rhat_df = pd.concat(all_bad_rhat_params)
    all_bad_rhat_summary_path = os.path.join(diagnostic_dir, "all_bad_group_rhat_summary.csv")
    all_bad_rhat_df.to_csv(all_bad_rhat_summary_path, encoding='utf-8-sig')
    print(f"All models' non-converged group-level parameters saved: {all_bad_rhat_summary_path}")

print("‚úÖ Batch model diagnostics & traceplot finished!\n")

# ========== ÈôÑÊ≥® ==========
# 1. ‰ªÖËØäÊñ≠Ôºå‰∏çÂÅöÂèÇÊï∞Êé®Êñ≠Âíå‰∏ªÁªìÊûúÂèØËßÜÂåñÔºà
# 2. ÊîØÊåÅSumÁºñÁ†ÅÂèÇÊï∞ÂêçÔºåÈúÄ‰∏é‰Ω†ÁöÑÊ®°ÂûãÂÖ¨Âºè‰øùÊåÅ‰∏ÄËá¥ÔºÅ
# 3. _diagnostics/ ‰∏ãËá™Âä®ÂΩíÊ°£ÂÖ®ÈÉ®ËØäÊñ≠csvÂíåtraceplot
# 4. Â¶ÇÊûúÈúÄÊîØÊåÅTreatmentÁºñÁ†ÅÔºåÂè™ÈúÄMAIN_EFFECTSÂèÇÊï∞ÂêçÊç¢‰∏∫TreatmentÊ†ºÂºèÂç≥ÂèØ
# 5. Âè™ÈúÄÂú®ËÆ∫ÊñáË°•ÂÖÖÊùêÊñôÂΩíÊ°£diagnosticsÔºå‰∏ªÂõæ/‰∏ªË°®Âú®BËÑöÊú¨Â§ÑÁêÜ„ÄÇ
```

```python
import os
import gc
import hddm
import logging
import pandas as pd
from tqdm import tqdm
import numpy as np

# ========== ÈÖçÁΩÆ ==========
BASE_OUTDIR = "model_results"
result_dir = os.path.join(BASE_OUTDIR, "_model_compare")
os.makedirs(result_dir, exist_ok=True)

# good_models ÂàóË°®ÈúÄÂíåÂÆûÈôÖÂêàÊ†ºÊ®°Âûã‰øùÊåÅ‰∏ÄËá¥ÔºÅ
good_models = [
    'v_main','v_interaction','a_main','a_interaction','full_model'
]

logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s %(levelname)s: %(message)s'
)
logger = logging.getLogger(__name__)

# ========== 1. ÊâπÈáèÂØºÂá∫ DICÔºåËá™Âä®Ê∏ÖÁêÜÂÜÖÂ≠ò ==========
def get_dic(model_name, model_dir):
    hddm_path = os.path.join(model_dir, f"{model_name}.hddm")
    if not os.path.exists(hddm_path):
        logger.warning(f"‚ùå {hddm_path} ‰∏çÂ≠òÂú®ÔºåË∑≥ËøáËØ•Ê®°Âûã„ÄÇ")
        return None
    try:
        logger.info(f"[{model_name}] Ê≠£Âú®Âä†ËΩΩÊ®°ÂûãÔºåËÆ°ÁÆó DIC ...")
        model = hddm.load(hddm_path)
        dic_val = model.dic
        del model
        gc.collect()
        logger.info(f"‚úÖ {model_name}: DIC = {dic_val:.2f}")
        return dic_val
    except Exception as e:
        logger.error(f"‚ùå {model_name} ËÆ°ÁÆó DIC Â§±Ë¥•: {e}")
        return None

# ========== 2. ÂèØÈÄâÔºöÊâπÈáèÂØºÂá∫ PPCÔºàÂ¶ÇÈúÄÂêéÈ™åÊãüÂêàÊ£ÄÈ™åÁ≠âÔºâ ==========
def export_ppc_infdata(model_name, model_dir):
    hddm_path = os.path.join(model_dir, f"{model_name}.hddm")
    nc_path = os.path.join(model_dir, f"{model_name}_ppc.nc")
    if not os.path.exists(hddm_path):
        logger.warning(f"‚ùå {hddm_path} ‰∏çÂ≠òÂú®ÔºåË∑≥ËøáPPC„ÄÇ")
        return False
    if os.path.exists(nc_path):
        logger.info(f"{nc_path} Â∑≤Â≠òÂú®ÔºåË∑≥Ëøá„ÄÇ")
        return True
    try:
        logger.info(f"[{model_name}] Ê≠£Âú®ÂØºÂá∫PPC infdata ...")
        model = hddm.load(hddm_path)
        infdata = model.to_infdata(ppc=True, loglike=False, save_name=nc_path)
        del model, infdata
        gc.collect()
        logger.info(f"‚úÖ {model_name}: PPC infdata Â∑≤ÂØºÂá∫„ÄÇ")
        return True
    except Exception as e:
        logger.error(f"‚ùå {model_name} ÂØºÂá∫PPCÂ§±Ë¥•: {e}")
        return False

# ========== 3. ‰∏ªÂæ™ÁéØÔºöÊâπÈáèËÆ°ÁÆó DICÔºåÂπ∂ÂèØÈÄâÂØºÂá∫PPC ==========
dic_results = []
for name in tqdm(good_models, desc="ÊâπÈáèËÆ°ÁÆóDIC/ÂèØÈÄâPPC"):
    model_dir = os.path.join(BASE_OUTDIR, name)
    dic_val = get_dic(name, model_dir)
    if dic_val is not None:
        dic_results.append({'model': name, 'DIC': dic_val})
        # Ëã•ÈúÄË¶ÅËá™Âä®ÊâπÈáèÂØºÂá∫PPCÔºåÂèñÊ∂à‰∏ã‰∏ÄË°åÊ≥®Èáä
        # export_ppc_infdata(name, model_dir)

# ========== 4. ËæìÂá∫Ê®°ÂûãÊéíÂ∫èÁªìÊûú ==========
if len(dic_results) >= 2:
    dic_df = pd.DataFrame(dic_results)
    dic_df.sort_values('DIC', inplace=True)
    dic_csv = os.path.join(result_dir, 'dic_compare.csv')
    dic_df.to_csv(dic_csv, index=False, encoding='utf-8-sig')
    logger.info(f"\nÊ®°ÂûãDICÊØîËæÉË°®Â∑≤‰øùÂ≠òËá≥: {dic_csv}\n")
    print(dic_df)

    # ËÆ°ÁÆóÊ®°ÂûãDIC weightÔºàËøë‰ººÊ¶ÇÁéáÔºåË∂äÂ∞èË∂äÂ•ΩÔºâ
    dic_vals = dic_df['DIC'].values
    dic_weights = np.exp(-0.5 * (dic_vals - np.min(dic_vals)))
    dic_weights = dic_weights / dic_weights.sum()
    dic_df['weight'] = dic_weights
    weight_csv = os.path.join(result_dir, 'model_dic_weights.csv')
    dic_df.to_csv(weight_csv, index=False, encoding='utf-8-sig')
    logger.info(f"Ê®°ÂûãDICÊùÉÈáçË°®Â∑≤‰øùÂ≠òËá≥: {weight_csv}")
    print(dic_df[['model','DIC','weight']])
    print("\n-- ËØ¥ÊòéÔºöweight‰∏∫Akaike weightÔºåÂèçÊò†Ê®°ÂûãË¢´Êï∞ÊçÆÊîØÊåÅÁöÑÊ¶ÇÁéáÂàÜÂ∏É„ÄÇ")
else:
    logger.warning("‚ö†Ô∏è Ê≤°ÊúâË∂≥Â§üÁöÑÊ®°ÂûãÂèØÊØîËæÉÔºàËá≥Â∞ë2‰∏™ÔºâÔºÅ")

logger.info("ÊµÅÁ®ãÂÖ®ÈÉ®ÁªìÊùü„ÄÇ")
```

üö©chunk 8: Ê†πÊçÆÊØîËæÉÁªìÊûúÂíåÁêÜËÆ∫ËøõË°åÂèØËßÜÂåñ

```python
import os
import pandas as pd
import arviz as az
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde

# ========== 1. BASIC CONFIGURATION ==========
BASE_OUTDIR = "model_results"
MODEL_NAMES = [
    'v_main', 'v_interaction', 'a_main', 'a_interaction', 'full_model'
]
PARAM_PREFIX = ['v', 'a', 'z', 't']
PARAM_PATTERNS = ["v_", "a_"]
ROPE = [-0.05, 0.05]
REF_VAL = 0

# Jeffreys evidence categories (JASP/standard)
BF10_THRESHOLDS = [1/100, 1/30, 1/10, 1/3, 1, 3, 10, 30, 100]
BF10_LEVELS = [
    "Extreme evidence for H0",
    "Very strong evidence for H0",
    "Strong evidence for H0",
    "Moderate evidence for H0",
    "Anecdotal evidence for H0",
    "Anecdotal evidence for H1",
    "Moderate evidence for H1",
    "Strong evidence for H1",
    "Very strong evidence for H1",
    "Extreme evidence for H1"
]

MAIN_EFFECTS = [
    "v_C(emotion, Sum)[S.aff]", "v_C(emotion, Sum)[S.dis]", "v_C(emotion, Sum)[S.dom]", "v_C(emotion, Sum)[S.enj]",
    "v_C(offer_type, Sum)[S.unfair]",
    "v_C(emotion, Sum)[S.aff]:C(offer_type, Sum)[S.unfair]",
    "v_C(emotion, Sum)[S.dis]:C(offer_type, Sum)[S.unfair]",
    "v_C(emotion, Sum)[S.dom]:C(offer_type, Sum)[S.unfair]",
    "v_C(emotion, Sum)[S.enj]:C(offer_type, Sum)[S.unfair]",
    "a_C(emotion, Sum)[S.aff]", "a_C(emotion, Sum)[S.dis]", "a_C(emotion, Sum)[S.dom]", "a_C(emotion, Sum)[S.enj]",
    "a_C(offer_type, Sum)[S.unfair]",
    "a_C(emotion, Sum)[S.aff]:C(offer_type, Sum)[S.unfair]",
    "a_C(emotion, Sum)[S.dis]:C(offer_type, Sum)[S.unfair]",
    "a_C(emotion, Sum)[S.dom]:C(offer_type, Sum)[S.unfair]",
    "a_C(emotion, Sum)[S.enj]:C(offer_type, Sum)[S.unfair]",
]

def ensure_dir(folder):
    if not os.path.exists(folder):
        os.makedirs(folder)

def get_significance_en(row):
    if row['hdi_2.5%'] > 0:
        return "Significantly Positive"
    elif row['hdi_97.5%'] < 0:
        return "Significantly Negative"
    else:
        return "Not Significant"

def get_rope_covered(row, rope=ROPE):
    if (row['hdi_2.5%'] > rope[1]) or (row['hdi_97.5%'] < rope[0]):
        return "Outside ROPE"
    elif (row['hdi_2.5%'] >= rope[0]) and (row['hdi_97.5%'] <= rope[1]):
        return "Inside ROPE"
    else:
        return "Overlap"

def format_float(val):
    try:
        if np.isnan(val) or np.isinf(val) or val == 0:
            return str(val)
        return f"{val:.6f}"
    except Exception:
        return str(val)

def classify_bf10(bf10):
    if np.isnan(bf10) or np.isinf(bf10) or bf10 == 0:
        return "Undefined"
    for i, th in enumerate(BF10_THRESHOLDS):
        if bf10 < th:
            return BF10_LEVELS[i]
    return BF10_LEVELS[-1]

def plot_prior_posterior_kde(ax, post_samples, prior_samples, xlim=(-4, 4), point=0.0, title=None, **plot_kwargs):
    x = np.linspace(xlim[0], xlim[1], 600)
    prior_kde = gaussian_kde(prior_samples)
    post_kde  = gaussian_kde(post_samples)
    ax.plot(x, prior_kde(x), '--', label='Prior')
    ax.plot(x, post_kde(x), '-', label='Posterior')
    ax.axvline(point, color='k', ls=':', alpha=0.5)
    # Standard: BF10 = prior@0 / posterior@0
    posterior0 = post_kde(point)[0]
    prior0 = prior_kde(point)[0]
    # Safeguard for both values extremely close to zero:
    if posterior0 < 1e-30 or prior0 < 1e-30:
        return np.nan
    bf10 = prior0 / posterior0
    ax.set_title(f"{title}\nBF‚ÇÅ‚ÇÄ={bf10:.2e}", fontsize=10)
    ax.set_xlabel(title, fontsize=9)
    ax.set_ylabel("Density", fontsize=9)
    ax.legend(fontsize=8)
    return bf10

def plot_kde_diag(param, post_samples, prior_samples, save_path=None, xlim=(-4,4)):
    x = np.linspace(xlim[0], xlim[1], 1000)
    prior_kde = gaussian_kde(prior_samples)
    post_kde = gaussian_kde(post_samples)
    plt.figure(figsize=(5,3))
    plt.plot(x, prior_kde(x), label='Prior', linestyle='--')
    plt.plot(x, post_kde(x), label='Posterior', linestyle='-')
    plt.axvline(0, color='k', linestyle=':', alpha=0.6)
    plt.scatter([0], [prior_kde(0)], c='red', label='Prior@0', zorder=5)
    plt.scatter([0], [post_kde(0)], c='blue', label='Posterior@0', zorder=5)
    plt.legend()
    plt.title(f"{param} Prior & Posterior KDE (0-point)")
    plt.xlabel('Parameter value')
    plt.ylabel('Density')
    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=150)
    plt.close()

def split_params(param_names):
    group, subj = [], []
    for p in param_names:
        if ('_subj.' in p) or ('_subj_' in p):
            subj.append(p)
        else:
            group.append(p)
    return group, subj

for model_name in MODEL_NAMES:
    print(f"\n=== Processing model: {model_name} ===")
    model_dir = os.path.join(BASE_OUTDIR, model_name)
    if not os.path.exists(model_dir):
        print(f"‚ùå Directory not found: {model_dir}, skipping.")
        continue

    nc_candidates = [
        f"{model_name}_grouponly.nc",
        f"{model_name}_loglike.nc",
        f"{model_name}.nc"
    ]
    for nc_file in nc_candidates:
        nc_path = os.path.join(model_dir, nc_file)
        if os.path.exists(nc_path):
            break
    else:
        print(f"‚ùå No nc file found for {model_name}, skipping.")
        continue

    inf = az.from_netcdf(nc_path)

    all_params = [v for v in inf.posterior.data_vars if any(v.startswith(p) for p in PARAM_PREFIX) and not v.endswith("_std")]
    group_params, subj_params = split_params(all_params)
    stat_params = [
        vn for patt in PARAM_PATTERNS for vn in inf.posterior.data_vars
        if patt in vn and not ('_subj.' in vn or '_subj_' in vn)
    ]
    stat_subj_params = [
        vn for patt in PARAM_PATTERNS for vn in inf.posterior.data_vars
        if patt in vn and ('_subj.' in vn or '_subj_' in vn)
    ]

    subdirs = {k: os.path.join(model_dir, k) for k in ["posterior", "forest", "ppc", "summary", "mainfig"]}
    for folder in subdirs.values():
        ensure_dir(folder)

    # === 1. Posterior & Forest plots for each main effect ===
    for param in MAIN_EFFECTS:
        if param in group_params:
            try:
                fig = az.plot_posterior(inf, var_names=[param], figsize=(8, 4), hdi_prob=0.95, show=False)
                plt.title(f"Posterior of {param}", fontsize=14)
                plt.axvline(0, ls=":", c="k", alpha=0.5)
                plt.tight_layout()
                plt.savefig(os.path.join(subdirs["mainfig"], f"{model_name}_posterior_{param}.tiff"), dpi=300, bbox_inches="tight")
                plt.close()
            except Exception as e:
                print(f"Posterior {param} failed: {e}")
            try:
                fig = az.plot_forest(inf, var_names=[param], combined=True, hdi_prob=0.95, show=False)
                plt.title(f"Forest: {param}")
                plt.axvline(0, ls=":", c="k", alpha=0.5)
                plt.tight_layout()
                plt.savefig(os.path.join(subdirs["mainfig"], f"{model_name}_forest_{param}.tiff"), dpi=300, bbox_inches="tight")
                plt.close()
            except Exception as e:
                print(f"Forest {param} failed: {e}")

    # === 2. Posterior All/Group Plots ===
    if len(group_params) > 0:
        try:
            fig = az.plot_posterior(inf, var_names=group_params, figsize=(max(6, len(group_params)*1.5), 3*len(group_params)), show=False)
            plt.suptitle(f"Posterior Distributions (group-level): {model_name}", fontsize=16)
            plt.tight_layout()
            plt.savefig(os.path.join(subdirs["posterior"], f"{model_name}_posterior_group_all.tiff"), dpi=300, bbox_inches="tight")
            plt.close()
        except Exception as e:
            print(f"Group-level Posterior failed: {e}")

    main_effects = [p for p in MAIN_EFFECTS if ":" not in p]
    interaction_effects = [p for p in MAIN_EFFECTS if ":" in p]
    if any([p in group_params for p in main_effects]):
        try:
            fig = az.plot_posterior(inf, var_names=[p for p in main_effects if p in group_params],
                                    figsize=(max(8, len(main_effects)*1.2), 3), show=False)
            plt.suptitle(f"Main Effects: {model_name}", fontsize=16)
            plt.axvline(0, ls=":", c="k", alpha=0.5)
            plt.tight_layout()
            plt.savefig(os.path.join(subdirs["mainfig"], f"{model_name}_main_effects_posterior.tiff"), dpi=300, bbox_inches="tight")
            plt.close()
        except Exception as e:
            print(f"Main effects posterior failed: {e}")
    if any([p in group_params for p in interaction_effects]):
        try:
            fig = az.plot_posterior(inf, var_names=[p for p in interaction_effects if p in group_params],
                                    figsize=(max(8, len(interaction_effects)*1.2), 3), show=False)
            plt.suptitle(f"Interaction Effects: {model_name}", fontsize=16)
            plt.axvline(0, ls=":", c="k", alpha=0.5)
            plt.tight_layout()
            plt.savefig(os.path.join(subdirs["mainfig"], f"{model_name}_interaction_effects_posterior.tiff"), dpi=300, bbox_inches="tight")
            plt.close()
        except Exception as e:
            print(f"Interaction effects posterior failed: {e}")

    if len(group_params) > 0:
        try:
            fig = az.plot_forest(inf, var_names=group_params, combined=True, hdi_prob=0.95, show=False)
            plt.title(f"Forest Plot (group-level): {model_name}")
            plt.axvline(0, ls=":", c="k", alpha=0.5)
            plt.tight_layout()
            plt.savefig(os.path.join(subdirs["forest"], f"{model_name}_forest_group.tiff"), dpi=300, bbox_inches="tight")
            plt.close()
        except Exception as e:
            print(f"Full forest plot failed: {e}")

    # === 3. Posterior Predictive Check (PPC) ===
    try:
        fig = az.plot_ppc(inf, group="posterior", figsize=(8, 4), show=False)
        plt.suptitle(f"PPC: {model_name}", fontsize=16)
        plt.tight_layout()
        plt.savefig(os.path.join(subdirs["ppc"], f"{model_name}_ppc.tiff"), dpi=300, bbox_inches="tight")
        plt.close()
    except Exception as e:
        print(f"PPC plot failed: {e}")

    # === 4. STATISTICAL INFERENCE (HDI/ROPE/BF10/KDE PLOTS) ===
    results_group = []
    extreme_bf10_rows = []
    for param in stat_params:
        try:
            summary = az.summary(inf, var_names=[param], hdi_prob=0.95)
            diag = az.summary(inf, var_names=[param], kind='diagnostics')
            summary['r_hat'] = diag['r_hat']
            summary['significance'] = summary.apply(get_significance_en, axis=1)
            summary['ROPE_covered'] = summary.apply(get_rope_covered, axis=1)
        except Exception as e:
            print(f"Summary failed for {param}: {e}")
            continue
        try:
            post_samples = inf.posterior[param].values.flatten()
            if hasattr(inf, "prior") and param in getattr(inf, "prior", {}):
                prior_samples = inf.prior[param].values.flatten()
            else:
                matched = [k for k in getattr(inf, "prior", {}).keys() if k[0] == param[0]]
                if matched:
                    prior_samples = inf.prior[matched[0]].values.flatten()
                else:
                    prior_samples = np.random.normal(0, 1, size=len(post_samples))
            bf10 = plot_prior_posterior_kde(
                plt.gca(), post_samples, prior_samples, xlim=(-3, 3), title=param
            )
            plt.close()
            save_path = os.path.join(subdirs["mainfig"], f"{model_name}_{param}_KDE_diag.png")
            plot_kde_diag(param, post_samples, prior_samples, save_path)
        except Exception as e:
            bf10 = np.nan
        summary['BF10'] = bf10
        summary['BF10_level'] = classify_bf10(bf10)
        results_group.append(summary)
        # Extreme value screening (for export or additional review)
        if classify_bf10(bf10) in [
            "Extreme evidence for H0", "Extreme evidence for H1",
            "Very strong evidence for H0", "Very strong evidence for H1"
        ]:
            extreme_bf10_rows.append(summary)

    # ===== Export full table =====
    if results_group:
        group_summary = pd.concat(results_group)
        group_summary.index.name = "param"
        group_summary.reset_index(inplace=True)
        group_summary['BF10'] = group_summary['BF10'].apply(format_float)
        group_summary_path = os.path.join(subdirs["summary"], f"group_level_summary_BF10.csv")
        group_summary.to_csv(group_summary_path, encoding='utf-8-sig', index=False)
        main_summary = group_summary[group_summary['param'].isin(MAIN_EFFECTS)]
        main_summary_path = os.path.join(subdirs["mainfig"], f"main_effects_group_level_summary_BF10.csv")
        main_summary.to_csv(main_summary_path, encoding='utf-8-sig', index=False)
        if extreme_bf10_rows:
            extreme_df = pd.concat(extreme_bf10_rows)
            extreme_path = os.path.join(subdirs["summary"], "extreme_BF10_params.xlsx")
            try:
                import openpyxl
            except ImportError:
                print("openpyxl not installed; cannot export to Excel. Use pip install openpyxl if needed.")
            else:
                extreme_df['BF10'] = extreme_df['BF10'].apply(format_float)
                extreme_df.to_excel(extreme_path, index=False)
                print(f"Extreme BF10 parameters exported: {extreme_path}")

        # ====== Write a separate README file ======
        description_txt = os.path.join(subdirs["summary"], "README_BF10.txt")
        with open(description_txt, "w", encoding="utf-8") as f:
            f.write(
                "BF10 (Bayes Factor): Computed as prior density at zero divided by posterior density at zero (Savage-Dickey ratio).\n"
                "BF10 > 1: evidence for the alternative hypothesis (effect present).\n"
                "BF10 < 1: evidence for the null hypothesis (no effect).\n"
                "Evidence levels (Jeffreys, 1961):\n"
                "    BF10 < 1/100: Extreme evidence for H0\n"
                "    1/100 ‚â§ BF10 < 1/30: Very strong evidence for H0\n"
                "    1/30 ‚â§ BF10 < 1/10: Strong evidence for H0\n"
                "    1/10 ‚â§ BF10 < 1/3: Moderate evidence for H0\n"
                "    1/3 ‚â§ BF10 < 1: Anecdotal evidence for H0\n"
                "    1 ‚â§ BF10 < 3: Anecdotal evidence for H1\n"
                "    3 ‚â§ BF10 < 10: Moderate evidence for H1\n"
                "    10 ‚â§ BF10 < 30: Strong evidence for H1\n"
                "    30 ‚â§ BF10 < 100: Very strong evidence for H1\n"
                "    BF10 ‚â• 100: Extreme evidence for H1\n"
                "All results use the JASP/BayesFactor default BF10 direction (prior/posterior at zero).\n"
                "If 'BF10_level' is 'Undefined', it means that the posterior and/or prior density at zero is numerically zero (or missing), "
                "which usually reflects extremely strong evidence for H1, but cannot be meaningfully quantified. Check diagnostic plots.\n"
            )
        print(f"Results and description exported for model: {model_name}")

print("\nüéâ All models have been processed and archived with standard BF10 (prior/posterior at 0) reporting!")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ËØªÂèñtraceÂêéÈ™åÊ†∑Êú¨Êñá‰ª∂
trace = pd.read_csv("model_results/full_model/full_model_trace.csv")

# Êò†Â∞Ñ
emotion_keys = ["dis", "dom", "neu", "aff", "enj"]
full_label = {
    'dis': 'Disgust',
    'dom': 'Dominance',
    'neu': 'Neutral',
    'aff': 'Affiliative',
    'enj': 'Reward'
}

emotion_param_map = {
    'dis': "v_C(emotion, Sum)[S.dis]",
    'dom': "v_C(emotion, Sum)[S.dom]",
    'neu': None,
    'aff': "v_C(emotion, Sum)[S.aff]",
    'enj': "v_C(emotion, Sum)[S.enj]"
}

interaction_param_map = {
    'dis': "v_C(emotion, Sum)[S.dis]:C(offer_type, Sum)[S.fair]",
    'dom': "v_C(emotion, Sum)[S.dom]:C(offer_type, Sum)[S.fair]",
    'neu': None,
    'aff': "v_C(emotion, Sum)[S.aff]:C(offer_type, Sum)[S.fair]",
    'enj': "v_C(emotion, Sum)[S.enj]:C(offer_type, Sum)[S.fair]"
}

v_intercept = trace['v_Intercept']
v_offer_fair = trace['v_C(offer_type, Sum)[S.fair]']

# È¢úËâ≤Êò†Â∞ÑÔºàÂèØË∞ÉÊï¥Ôºâ
emotion_colors = {
    'Disgust': "#8c510a",
    'Dominance': "#e08214",
    'Neutral': "#bababa",
    'Affiliative': "#39E04F",
    'Reward': "#d73027"
}

plt.figure(figsize=(14,7))

# ÂÖàËÆæÁΩÆÈ£éÊ†º
sns.set_style("whitegrid")
sns.set_context("notebook", font_scale=1.2)

for emo in emotion_keys:
    v_emo_samples = trace[emotion_param_map[emo]] if emotion_param_map[emo] else 0
    v_interact_samples = trace[interaction_param_map[emo]] if interaction_param_map[emo] else 0

    drift_unfair = v_intercept + v_emo_samples
    drift_fair = v_intercept + v_emo_samples + v_offer_fair + v_interact_samples

    # KDE‰º∞ËÆ°ÔºöÂÖ¨Âπ≥Áî®ÂÆûÁ∫øÔºåÈùûÂÖ¨Âπ≥Áî®ËôöÁ∫øÔºåÊõ¥Ê∏ÖÊô∞Âå∫ÂàÜ
    sns.kdeplot(
        drift_unfair, bw_adjust=0.5, fill=True, alpha=0.25,
        label=f"{full_label[emo]} - Unfair",
        color=emotion_colors[full_label[emo]], linestyle='-', linewidth=1.3
    )
    sns.kdeplot(
        drift_fair, bw_adjust=0.5, fill=True, alpha=0.5,
        label=f"{full_label[emo]} - Fair",
        color=emotion_colors[full_label[emo]], linestyle='--', linewidth=2
    )

plt.axvline(0, color='gray', linestyle=':', linewidth=1)
plt.xlabel('Drift Rate (v)')
plt.ylabel('Density')
plt.title('Posterior Distributions of Drift Rate (v) by Proposer Emotion √ó Offer Type')
plt.legend(title='Condition', bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.tight_layout(rect=[0, 0, 0.85, 1])

plt.savefig("drift_rate_posterior.tif", dpi=600, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. ËØªÂèñtraceÂêéÈ™åÊ†∑Êú¨Êñá‰ª∂ÔºåÊõøÊç¢Ë∑ØÂæÑ‰∏∫‰Ω†ÁöÑÊñá‰ª∂Ë∑ØÂæÑ
trace = pd.read_csv("model_results/full_model/full_model_trace.csv")

# 2. ÂèÇÊï∞Êò†Â∞ÑÔºàvÁöÑÂõûÂΩíÁ≥ªÊï∞Ôºâ
params = {
    'v_Intercept': 'Intercept',
    'v_C(emotion, Sum)[S.dis]': 'Disgust',
    'v_C(emotion, Sum)[S.dom]': 'Dominance',
    'v_C(emotion, Sum)[S.aff]': 'Affiliative',
    'v_C(emotion, Sum)[S.enj]': 'Reward',
    'v_C(offer_type, Sum)[S.fair]': 'Fair Offer',
    'v_C(emotion, Sum)[S.dis]:C(offer_type, Sum)[S.fair]': 'Disgust √ó Fair',
    'v_C(emotion, Sum)[S.dom]:C(offer_type, Sum)[S.fair]': 'Dominance √ó Fair',
    'v_C(emotion, Sum)[S.aff]:C(offer_type, Sum)[S.fair]': 'Affiliative √ó Fair',
    'v_C(emotion, Sum)[S.enj]:C(offer_type, Sum)[S.fair]': 'Reward √ó Fair'
    # neuÂíåneu√ófair‰Ωú‰∏∫Âü∫Á∫øÔºå‰∏çÂåÖÂê´
}

# 3. ‰ªétraceÊèêÂèñÊï∞ÊçÆÔºåÊï¥ÁêÜ‰∏∫ÈïøÊ†ºÂºèÊñπ‰æøÁªòÂõæ
rows = []
for param_code, param_label in params.items():
    if param_code in trace.columns:
        samples = trace[param_code]
        df = pd.DataFrame({
            'Parameter': param_label,
            'Value': samples
        })
        rows.append(df)

df_long = pd.concat(rows, ignore_index=True)

# 4. ËÆæÁΩÆÁªòÂõæÈ£éÊ†º
plt.figure(figsize=(12, 7))
sns.violinplot(
    data=df_long,
    y='Parameter',
    x='Value',
    palette='muted',
    inner='quartile',
    scale='width',
    cut=0
)

plt.axvline(0, color='gray', linestyle='--')
plt.title('Posterior Distributions of Regression Coefficients for Drift Rate (v)')
plt.xlabel('Regression Coefficient')
plt.ylabel('')
plt.tight_layout()

# 5. ‰øùÂ≠òÈ´òÂàÜËæ®ÁéáÂõæÁâá
plt.savefig("regression_coefficients_violin.tif", dpi=400, bbox_inches='tight')
plt.show()
```
